Looking at past three year's data on Unicorn's registration year and Unicorn's sector.

```python

from matplotlib.patches import ConnectionPatch

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(9, 5))
fig.subplots_adjust(wspace=0)

year_frequency = df['Entry_year'].value_counts()[:3]
overall_ratios = year_frequency
explode = [0, 0.1, 0]

angle = -150 * overall_ratios[0]
wedges, *_ = ax1.pie(overall_ratios, labels=year_frequency.index, autopct='%1.1f%%', startangle=angle)


sector_frequency = df['Sector'].value_counts()[:3]
sector_ratios = sector_frequency
sector_labels = sector_frequency.index
bottom = 0
width = .2

for j, (height, label) in enumerate(reversed(list(zip(sector_ratios, sector_labels)))):
    bc = ax2.bar(0, height, width, bottom=bottom, color='C0', label=label, alpha=0.1 + 0.25 * j)
    ax2.bar_label(bc, labels=[f"{height:.0f}"], label_type='center')
    bottom += height

ax2.set_title('Sector')
ax2.legend()
ax2.axis('off')
ax2.set_xlim(-2.5 * width, 2.5 * width)

theta1, theta2 = wedges[0].theta1, wedges[0].theta2
center, r = wedges[0].center, wedges[0].r
bar_height = sum(sector_ratios)

x = r * np.cos(np.pi / 180 * theta2) + center[0]
y = r * np.sin(np.pi / 180 * theta2) + center[1]
con = ConnectionPatch(xyA=(-width / 2, bar_height), coordsA=ax2.transData,
                      xyB=(x, y), coordsB=ax1.transData)
con.set_color([0, 0, 0])
con.set_linewidth(4)
ax2.add_artist(con)

x = r * np.cos(np.pi / 180 * theta1) + center[0]
y = r * np.sin(np.pi / 180 * theta1) + center[1]
con = ConnectionPatch(xyA=(-width / 2, 0), coordsA=ax2.transData,
                      xyB=(x, y), coordsB=ax1.transData)
con.set_color([0, 0, 0])
ax2.add_artist(con)
con.set_linewidth(4)

plt.show()
```

Year 2021 had the heighest number of unicorn's registered. Top 3 sectors for Unicorn's registered at year 2021 are E-commerce, marketplace-used cars, edtech.
